```python
import pandas as pd
import numpy as np
```

```python
r_cols=['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movie_dataset=pd.read_csv('ml-100k/u.item',sep='|',names=r_cols,encoding='latin-1')
```

```python
movie_vector=movie_dataset[['Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
```

```python
movie_vector.index=range(1,1683)
movie_vector=movie_vector.T
movie_vector
```

```python
r_cols=['userID','movieID','ratings','timestamp']
user_dataset=pd.read_csv('ml-100k/u.data',sep='\t',names=r_cols,encoding='latin-1')
user_dataset
```

```python
user_vector=[]
for user in range(1,944):
    specuser_movieid=user_dataset[user_dataset['userID']==user][['movieID']] #Extracts movieID column and specuser_movieis also becomes dataframe with movieID column
    user_spec=[]
    for movieid in specuser_movieid['movieID']:
        user_spec.append(movie_vector[movieid].values)
    user_specdf=pd.DataFrame(user_spec,columns=range(1,19))
    user_specdf
    user_vec=[]                   
    for moviecategory in range(1,19):
        sum_val=sum(user_specdf[moviecategory])
        user_vec.append(sum_val/len(user_specdf))
    user_vector.append(user_vec)
user_vectdf=pd.DataFrame(user_vector)
user_vectdf
```

```python
user_vectdf.shape
```

```python
movie_vector.shape
```

```python
content_vector=np.dot(user_vectdf,movie_vector)
content_vector
```

```python
content_vectors=pd.DataFrame(content_vector)
content_vectors=content_vectors.T
content_vectors
```

```python
#getting the movies watched by specfic user thatis movie id's
user_watched=user_dataset[user_dataset['userID']==4]['movieID'] #now movieID's will extracted as series not as dataframe because single bracket[] not[[]]
user_watched
```

```python
spec_val=content_vectors[content_vectors[4]>(content_vectors[4].max()-0.3)][2]  
spec_val
```

```python
recommended_movies=[]
for movieid in spec_val.index:
    if movieid in user_watched:
        pass
    else:
        recommended_movies.append(movieid)
recommended_movies
    
```
